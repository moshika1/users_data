```python
import requests 
import csv 
```

```python
response = requests.get('https://jsonplaceholder.typicode.com/users')
response
```

```python
users = response.json()
users
```

```python
with open('users_data.csv' , 'w' , newline='') as f:
    writer = csv.DictWriter(f , fieldnames=['name' , 'username' , 'email' , 'street' , 'suite' , 'city' , 'zipcode' , 'geo_lat' , 'geo_lng'])
    writer.writeheader() 
    
    for users_data in users:
        name = users_data['name']
        username = users_data['username']
        email = users_data['email']
        street = users_data['address']['street']
        suite = users_data['address']['suite']
        city = users_data['address']['city']
        zipcode = users_data['address']['zipcode']
        geo_lat = users_data['address']['geo']['lat']
        geo_lng = users_data['address']['geo']['lng']
        
        writer.writerow({
             'name' : name ,
             'username' : username ,
             'email' : email ,
             'street' : street ,
             'suite' : suite ,
             'city' : city ,
             'zipcode' : zipcode ,
              'geo_lat' : geo_lat , 
               'geo_lng' : geo_lng
                })
```
